```python


import json
import matplotlib.pyplot as plt

with open("/home/james/ebi_monarch_xspecies_summary.json", "r") as f:
    sums = json.load(f)

type_to_count = sums['displaytypes']
sorted_types = sorted(type_to_count.items(), key=lambda item: item[1]['count'], reverse=True)

full_total_count = sum(item[1]['count'] for item in sorted_types)

top_10_types = sorted_types[:10]
labels = [item[0] for item in top_10_types]
sizes = [item[1]['count'] for item in top_10_types]

percentages = [(count / full_total_count) * 100 for count in sizes]
percentage_labels = [f"{label}: {p:.1f}%" for label, p in zip(labels, percentages)]

fig, ax = plt.subplots()

plt.text(0, 1.25, f"Top 10 types of {full_total_count:,} node->type assignments", ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.subplots_adjust(top=0.85)

ax.pie(sizes, labels=percentage_labels, startangle=90)
ax.axis('equal')

plt.show()

```
